```python
role_mapping = {
    'data scientist': 'Data Scientist',
    'software engineer': 'Software Engineer',
    'product manager': 'Product Manager',
    'data engineer': 'Data Engineer',
    'data analyst': 'Data Analyst',
    'ui engineer': 'UI Engineer',
    'ui designer': 'UI Designer',
    'AI Engineer': 'AI Researcher',
    'HR Specialist': 'Human Resources Specialist',
    'Cybersecurity Specialist': 'Cybersecurity Analyst',
    'UI/UX Designer': 'UI Designer',
    'Graphic Designer': 'UI Designer',
    'Network Engineer': 'System Administrator',
    'Software Developer': 'Software Engineer'
}

df['Role'] = df['Role'].replace(role_mapping)
```

```python
df['Role'].value_counts()
```

```python
df[['Role', 'Resume']].to_csv('data/cleaned_dataset.csv', index=False)
```
